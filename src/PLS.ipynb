```python
#importation des bibliothéques
from pyspectra.readers.read_spc import read_spc_dir, read_spc
from matplotlib import pyplot as plt
import pandas as pd 
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import cross_val_predict
from scipy.stats import linregress
```

```python
#Récupération des données d'entraînement sous forme d'un DataFrame 
from data_wrapper import *

df1=data_wrapper_j1()
df2=data_wrapper_j2()
df4=data_wrapper_j4()

df12=pd.concat([df1,df2],axis=0)
df124=pd.concat([df12,df4])

df=df124
df.describe()
```

## Construction et entraînement du modèle PLS pour le Glucose

```python
X_PLS_glu=np.array(df[df.keys()[7:]].values) # on prend toutes les données du spectre
Y_PLS_glu=df['Glucose'].tolist()
np.shape(X_PLS_glu)
```

```python
#Construction et entraînement du modèle PLS
nb_composantes=[k for k in range(1,14)]
mse_opt_glu=[]

for i in nb_composantes : 
    pls_opt_glu = PLSRegression(n_components=i)
    pls_opt_glu.fit(X_PLS_glu, Y_PLS_glu)
    # Cross-validation
    y_cv_opt_glu = cross_val_predict(pls_opt_glu, X_PLS_glu, Y_PLS_glu, cv=10)
    mse_opt_glu.append(mean_squared_error(Y_PLS_glu, y_cv_opt_glu))
```

```python
#Graphe du MSE en fonction du nombre de composantes
plt.plot(nb_composantes, np.array(mse_opt_glu), mfc='blue')
plt.xlabel('Number of PLS components')
plt.ylabel('MSE')
plt.title('PLS pour le glucose')
plt.xlim(left=-1)
plt.show()
```

```python
#Nombre optimal de composantes 
nb_optimal_glu=nb_composantes[mse_opt_glu.index(min(mse_opt_glu))]
print("Nb de composantes optimal : ",nb_optimal_glu)
```

```python
#Tracé
pls_glu = PLSRegression(n_components = nb_optimal_glu)
pls_glu.fit(X_PLS_glu, Y_PLS_glu)
y_c_glu = pls_opt_glu.predict(X_PLS_glu)
y_cv_glu = cross_val_predict(pls_glu, X_PLS_glu, Y_PLS_glu, cv=10)

# Calculate scores for calibration and cross-validation
score_c_glu = r2_score(Y_PLS_glu, y_c_glu)
score_cv_glu = r2_score(Y_PLS_glu, y_cv_glu)
print("score_cv_glu : ", score_cv_glu)
print("score_c_glu :",score_c_glu)

# Fit a line to the CV vs response
z = np.polyfit(Y_PLS_glu,y_c_glu,deg=1)
fig, ax = plt.subplots(figsize=(9, 5))
ax.scatter(y_c_glu, Y_PLS_glu, c='red', edgecolors='k')

#Plot the best fit line
ax.plot(np.polyval(z,Y_PLS_glu), Y_PLS_glu, c='blue', linewidth=1)
plt.title('$R^{2}$ (CV): '+ str(score_cv_glu))
plt.xlabel('Concentration en glucose prédite')
plt.ylabel('Concentration en glucose mesurée')
plt.show()
```

## Entraînement pour éthanol

```python
X_PLS_eth=np.array(df[df.keys()[7:]].values) # on prend les mêmes données d'entraînement que pour l gluc
Y_PLS_eth=df['Ethanol'].tolist()
np.shape(Y_PLS_eth),np.shape(X_PLS_eth)
```

```python
# Construction et entraînement du modèle PLS
nb_composantes=[k for k in range(1,14)]
mse_opt_eth=[]

for i in nb_composantes : 
    pls_opt_eth = PLSRegression(n_components=i)
    pls_opt_eth.fit(X_PLS_eth, Y_PLS_eth)
    # Cross-validation
    y_cv_opt_eth = cross_val_predict(pls_opt_eth, X_PLS_eth, Y_PLS_eth, cv=10)
    mse_opt_eth.append(mean_squared_error(Y_PLS_eth, y_cv_opt_eth))
```
